### Create/Train Logistic Regressor

```python
# Create and train the Logistic Regression model
# 'solver' is chosen for good performance on small datasets and 'random_state' for reproducibility
log_reg = LogisticRegression(solver='liblinear', random_state=42)
log_reg.fit(x_train_scaled, np.squeeze(y_train))

# Make predictions on the test set
y_pred = log_reg.predict(x_test_scaled)
y_pred_proba = log_reg.predict_proba(x_test_scaled)[:, 1] # Probability of the positive class (1)
```

## Random Forest

### Default Random Forest Model Setup (fixed params)

```python
# Instantiate the Random Forest Classifier
# n_estimators: Number of trees in the forest. More trees generally means better performance but slower.
# random_state: For reproducibility.
# class_weight: 'balanced' is crucial for imbalanced datasets (like your diabetes data).
#               It adjusts weights inversely proportional to class frequencies.
rf_model_default = RandomForestClassifier(
    n_estimators=100,      # Default is 100 trees
    random_state=42,
    n_jobs=-1,             # Use all available CPU cores for parallelization
    class_weight='balanced' # Helps with imbalanced classes
)
```

### Default Random Forest Model Evaluation

```python
#
# Metrics for default model setup
#
print("Model Eval")
y_pred_default = rf_model_default.predict(x_test_scaled)
y_pred_proba_default = rf_model_default.predict_proba(x_test_scaled)[:, 1] # Probabilities for the positive class

print(f"Accuracy: {accuracy_score(y_test, y_pred_default):.4f}")
print("\nClassification Report:\n", classification_report(y_test, y_pred_default))
print(f"ROC-AUC Score: {roc_auc_score(y_test, y_pred_proba_default):.4f}")

# You can also plot Precision-Recall curve, which is often more informative for imbalanced data
precision, recall, _ = precision_recall_curve(y_test, y_pred_proba_default)
pr_auc = auc(recall, precision)
print(f"Precision-Recall AUC: {pr_auc:.4f}")
```

### Random Forest Hyperparameter Tune

```python
#
# Use Gridsearch to find the best hyperparameters for Random Forest
#


print("Configuring GridSearchCV for Random Forest (Hyperparameter Tuning)")


#previously computed best parameters (to skip over gridsearch for debugging) done once
#Best parameters found by GridSearchCV: {'class_weight': 'balanced', 'max_depth': 10, 'max_features': 'sqrt',
#'min_samples_leaf': 4, 'n_estimators': 200}
#Best cross-validation score (ROC-AUC): 0.8216


# Define the parameter grid
param_grid = {
    'n_estimators': [50, 100, 200], # Number of trees in the forest
    'max_features': ['sqrt', 'log2'], # Number of features to consider at each split
    'max_depth': [5, 10,200], #, None], # Maximum depth of the tree (None means unlimited)
    'min_samples_leaf': [1, 2, 4], # Minimum number of samples required to be at a leaf node
    'class_weight': ['balanced'] # Keep balanced for imbalanced data
}

# Instantiate GridSearchCV
# scoring='roc_auc' is often a good choice for imbalanced binary classification
grid_search_rf = GridSearchCV(
    estimator=RandomForestClassifier(random_state=42,n_jobs=-1),
    param_grid=param_grid,
    cv=3,                 # Number of cross-validation folds
    scoring='roc_auc',    # Optimize for ROC-AUC as it's robust to imbalance
    n_jobs=-1,            # Use all CPU cores for parallelization
    verbose=1             # Show progress
)
```

### Grid Fit

```python
#
# Train the random forest model using the gridsearch for multiple params
# -- this takes a long time to train

#start timer for this block
start_time = time.time()


print("Fitting Grid Search (this may take some time)...")
grid_search_rf.fit(x_train_scaled, np.squeeze(y_train))
print("Grid Search complete.")

# 6a. Display Best Parameters and Score
print(f"\nBest parameters found: {grid_search_rf.best_params_}")
print(f"Best cross-validation score (ROC-AUC): {grid_search_rf.best_score_:.4f}")


#Show time for execution
end_time = time.time()
execution_time = end_time - start_time
print(f"Execution time: {execution_time} seconds")
```

### Grid Evaluation

```python
#
# Evaluate the model with best parameters on the scaled test data
#


print("Final Model Evaluation (Best Random Forest from GridSearchCV)")
best_rf_model = grid_search_rf.best_estimator_
y_pred_best_rf = best_rf_model.predict(x_test_scaled)
y_pred_proba_best_rf = best_rf_model.predict_proba(x_test_scaled)[:, 1]

print(f"Test Set Accuracy: {accuracy_score(y_test, y_pred_best_rf):.4f}")
print("\nClassification Report (Test Set):\n", classification_report(y_test, y_pred_best_rf))
print(f"Test Set ROC-AUC Score: {roc_auc_score(y_test, y_pred_proba_best_rf):.4f}")
precision_best, recall_best, _ = precision_recall_curve(y_test, y_pred_proba_best_rf)
pr_auc_best = auc(recall_best, precision_best)
print(f"Test Set Precision-Recall AUC: {pr_auc_best:.4f}")
```

### Data Information

**NOTE: Results taken from a sample run and may not exactly match a re-run of the script.**

  The data required scaling to get the features within a range acceptable for the models. No NaNs were found in either the UCI import or the raw csv data. The CSV file did require putting the pre-diabetes (1) and diabetes (2) classifiers together, to match the UCI imported data. There are 253680 records, with 213703 records assigned to the no-diabetes class (0) and 39977 assigned to the diabetes class (1).
  The CSV data set includes 3 classes; no-diabetes, pre-diabetes, and diabetes. The python imported dataset uses just no-diabetes and diabetes. To make the two methods identical, the CSV contents had pre-diabetes and diabetes cases together under the diabetes label.

  Class sizes are unbalanced, with no-diabetes (0) accounting for roughly 84% of the dataset. The no-diabetes and diabetes classes together, made up the rest. Doing a correlation between all of the features, it was found that General Health, Physical Health, DiffWalk,Income, and Education were the most highly correlated features. It is odd that features, such as BMI, age, and sex were not correlated higher.

  Common scaling was needed to account for some features that utilized larger numbers than the rest of the data, such as BMI. 













  Since this was tabular data, starting out with the more basic machine learning methods would be a good start. For that, logistic regression, Random Forest, and XGBoost would be used. A Support Vector Machine would be used to try out a deep machine learning method. Later, a Deep Neural Net was added to try out another method.  


### Logistic Classifier


  
  Using one of the more basic machine learning methods, linear regression, for this tabular data provided mixed results. With data that was scaled, the majority of the records fell under the 'no-diabetes' category (label=0). The classes are unbalanced, as the number of patients in the 'diabetes' category numbered 7069, whereas the patients reporting as 'no-diabetes' was at 43667. This test split gave us a good accuracy of ~85%, which means we predicted right 84.81% of the time on our test set. There is an issue with this though. For the data set, roughly 86% of the cases are in the 'no-diabetes' category, so if a model always predicted 'no-diabetes', it would be expected to get 86% of the predictions correct anyways (we actually did slightly worse than this). We will need to rely on the precision instead for this model. 

  
  The low precision, about 55%, meant that the model would give false positives almost half of the time. The recall value, 0.1910, signifies that only 19% of the true diabates predictions are correct. means the predictions only get about 16% of the true diabetes predictions right. f1 is a more balanced score, which can help with unbalanced data, but it was also low here. ## Random Forest

  To build upon the linear (logistic) regression methods, random forest modeling was used. The grouping of multiple decision trees is hoped to provide better results than linear regression models for our binary classifcation set. As trees, our branch level should provide a method to find non-linear patterns that the logistic regressor may not be able to detect. 


-  Default Hyperparameters

  The first random forest model used was built with a sample of default parameters. This included the default of 100 max trees and using class weighting balancing to compensate for the highly unbalanced dataset we have. The weights for the smaller class (diabetes) should be weighed more heavily than those given to the larger class (no-diabetes).

  The accuracy returned by the default random forest model was 84%, which is likely due to the overwhelming number of no-diabetes values present. Recall for no-diabetes was excellent, at 96%, but the diabetes recall was 20%. This only catches 20% of the responders that actually have diabetes (the rest are given false negatives). The mixed metric, f1-score, is sufficiently low for the diabetes cases to show that this model does not perform well, especially for a health related function. 
  
  Trying to optimize the method was accomplished by using the GridSearch method to rotate through different hyperparameters to find the best fit for our needs.

## Support Vector Machine

 Support Vector Machines are generally good at drawing boundaries for classification tasks. With our feature set and the hardware limitations of the current computer, it was decided to slim the feature set down to a smaller set of features. PCA was used to create a set of 6 features for this. Creating PCA features using variance level left too many features, so a hard parameter of 6 was used to force it a smaller number of features. 

 Training the SVM (SVC for classification) required **a lot** of computing resources and did not seem to have built-in GPU capabilities. This meant each run, even with a reduced number of features could take more than 3 hours (even on Google Colab's setup). Hardware constraints required keeping the SVM testing to a minimum as a new training would require excessive time to complete. 

  One solid run of SVM was completed and it produced a very high precision for the no-diabetes test cases, with 92% precision. For detecting if someone doesn't have diabetes, SVM produced a very good model. Likewise, recall was 71% for the no-diabetes predictions. This is in-line with the other models. When predicting diabetes cases, the precision of 30% severely hampers this model for that class. This low precision will cause a significant number of false positives to be reported. The differences between the precisions for the classes likely pushed the F1-score low. 

  Using PCA to reduce our training and test data may have removed some interaction between features that was picked up by other methods using the full feature set. Reducing features to something above 6 may provide better model metrics. Like all of the models, so far, the class imbalance heavily favors detecting no-diabetes instead of diabetes. The ROC curve, while better than random guesses, would need to be improved for overall health-related usage.

## Dense Neural Net

  With the other methods having difficulties with an unbalanced dataset, a neural net seemed like the next, easiest step. The standard layers were used, with the addition of a 0.2 dropout layer sandwiched between the hidden layers. 
  
  With a precision of 0.94, the DNN was able to successfully predict no-diabetes cases 94% of the time, which improves upon just guessing class 0. When predicting no-diabetes on actual diabetes cases, a recall of 0.72 shows that the model will correctly predict 72% of the actual no-diabetes cases. When predicting the diabetes cases, a precision of 0.33 means it will only correctly predict the actual diabetes cases 33% of the time. That leaves a lot of false positives for predictions. A recall of 0.76 is an improvement, which properly identifies 76% of the actual diabetes cases. 

  The high positive recall makes this a better-than-average diabetes detector for those patients having diabetes. The excess number of false positives for diabetes predictions would be a serious issue for using this for human needs. For detecting cases where the prediction is no-diabetes on cases where they don't actually have diabetes, this model has some merit, but the false positive problem would cause too many issues. 

  The training on DNNs can be optimized heavily, but due to time concerns, a relatively simple set of layers was used, but with enough layer setup to capture any of the more complicated features for output. During training, the loss did not drastically improve throughout the allotted 100 epochs. It could be that the learning rate was too low or too high or even that the number of ephocs to train did not allow it time to converge to something better. It was felt that keeping the full feature set with minimum training settings was favorable to reducing features and allocating more training resources.
